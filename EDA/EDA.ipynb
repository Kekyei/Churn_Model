## 5. Hypothesis investigation

Now that we have explored the data, it's time to investigate whether price sensitivity has some influence on churn. First we need to define exactly what is price sensitivity.

    > Since we have the consumption data for each of the companies for the year of 2015, we will create new features to measure "price sensitivity" using the average of the year, the last 6 months and the last 3 months

1. transform dates into dataframes
2. find the average of gas price AND electricity price for the whole year...
3. find the average of gas price AND electricity price for the first 3 months..TRY AND FIND THE AVERAGE OF BOTH THE SAME TIME
4. find the average of gas price AND electricity price for the last 6 months..RY AND FIND THE AVERAGE OF BOTH THE SAME TIME
5. find the total price of averages
6. merge all the data sets with churn
7. find the correlations with all of them

```python
#Converting all dates to datetime dataframes
client_data["date_activ"] = pd.to_datetime(client_data["date_activ"],format = "%Y-%m-%d")
client_data["date_end"] = pd.to_datetime(client_data["date_end"],format = "%Y-%m-%d")
client_data["date_modif_prod"] = pd.to_datetime(client_data["date_modif_prod"],format = "%Y-%m-%d")
client_data["date_renewal"] = pd.to_datetime(client_data["date_renewal"],format = "%Y-%m-%d")
price_data["price_date"] = pd.to_datetime(price_data["price_date"],format = "%Y-%m-%d")
price_data
```

```python
#To find out the behaviour of prices over the year, I'm finding out the mean prices for the whole year,after 3 months and after 6 months
year = price_data.groupby("id").mean().reset_index()
year_3m = price_data[price_data["price_date"]<"2015-04-01"].groupby("id").mean().reset_index() #first 3 months
year_6m = price_data[price_data["price_date"]>"2015-06-01"].groupby("id").mean().reset_index() #last 6 months
year
```

```python
# Comnbine into single dataframe and rename them
year = year.rename(
    index=str,
    columns={
        "price_off_peak_var": "mean_price_off_peak_var",
        "price_peak_var": "mean_price_peak_var",
        "price_mid_peak_var": "mean_price_mid_peak_var",
        "price_off_peak_fix": "mean_price_off_peak_fix",
        "price_peak_fix": "mean_price_peak_fix",
        "price_mid_peak_fix": "mean_price_mid_peak_fix"
    }
)
#This is also an easier way to rename the columns
year_3m.columns = ['id', 'mean3m_price_off_peak_var', 'mean3m_price_peak_var',
                   'mean3m_price_mid_peak_var', 'mean3m_price_off_peak_fix',
                   'mean3m_price_peak_fix', 'mean3m_price_mid_peak_fix']
year_6m.columns = ['id', 'mean6m_price_off_peak_var', 'mean6m_price_peak_var',
                   'mean6m_price_mid_peak_var', 'mean6m_price_off_peak_fix',
                   'mean6m_price_peak_fix', 'mean6m_price_mid_peak_fix']
year #renamed data frame
```

```python

#Finding out the average price a company paid for BOTH energy and electricity over the whole year and adding the values to the dataframe
year["mean_year_price_p1"] = year["mean_price_peak_fix"] + year["mean_price_peak_var"]
year["mean_year_price_p2"] = year["mean_price_off_peak_fix"] + year["mean_price_off_peak_var"]
year["mean_year_price_p3"] = year["mean_price_mid_peak_fix"] + year["mean_price_mid_peak_var"]
#Finding out the average price a company paid for BOTH energy and electricity after  months and adding the values to the dataframe
year_3m["mean3m_year_price_p1"] = year_3m["mean3m_price_peak_fix"] + year_3m["mean3m_price_peak_var"]
year_3m["mean3m_year_price_p2"] = year_3m["mean3m_price_off_peak_fix"] + year_3m["mean3m_price_off_peak_var"]
year_3m["mean3m_year_price_p3"] = year_3m["mean3m_price_mid_peak_fix"] + year_3m["mean3m_price_mid_peak_var"]
#Finding out the average price a company paid for BOTH energy and electricity after 6 months and adding the values to the dataframe
year_6m["mean6m_year_price_p1"] = year_6m["mean6m_price_peak_fix"] + year_6m["mean6m_price_peak_var"]
year_6m["mean6m_year_price_p2"] = year_6m["mean6m_price_off_peak_fix"] + year_6m["mean6m_price_off_peak_var"]
year_6m["mean6m_year_price_p3"] = year_6m["mean6m_price_mid_peak_fix"] + year_6m["mean6m_price_mid_peak_var"]
year_6m
```

```python
#Adding all the averages into one dataframe
price_features = pd.merge(year,year_6m,on = "id")
price_features = pd.merge(price_features,year_3m,on = "id")
price_features
```

```python
#Defining price sensitivity to churn
price_sensitivity = pd.merge(price_features,client_data[["churn","id"]],on = "id")
price_sensitivity
```

```python
# Finding out how strong price senstivity is related to a company churning
corr = price_sensitivity.corr()
plt.figure(figsize =(15,15))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values,annot=True,annot_kws={'size':10})
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

From the correlation plot, it shows a higher magnitude of correlation between other price sensitivity variables, however overall the correlation with churn is very low. This indicates that there is a weak linear relationship between price sensitity and churn. This suggests that for price sensitivity to be a major driver for predicting churn, we may need to engineer the feature differently.

```python
merged_data = pd.merge(price_sensitivity,client_data.drop(columns = ["churn"]),on="id")
```
